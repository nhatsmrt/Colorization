```python
from Source import \
    ClassColorNet, \
    create_data_places
from pathlib import Path

from sklearn.model_selection import train_test_split
```

```python
# DEFINE PATHS
d = str(Path().resolve())
data_path = d + "/data/images/train/"
export_path = d + "/Test Images/"
weight_save_path = d + "/Weights/ClassColorNet/1/"


# DEFINE HYPERPARAMETERS
n_epoch = 100
N_IMG_MAX_PER_CLASS = 200
N_CLASS_MAX = 40
class_weight = 1
patience = 10
```

```python
# Create dataset:
X, ab, y, n_classes = create_data_places(
    data_path,
    N_IMG_MAX_PER_CLASS = N_IMG_MAX_PER_CLASS,
    N_CLASS_MAX = N_CLASS_MAX
)

X_train_val, X_test, ab_train_val, ab_test, y_train_val, y_test = train_test_split(X, ab, y, test_size = 0.2)
X_train, X_val, ab_train, ab_val, y_train, y_val = train_test_split(X_train_val, ab_train_val, y_train_val, test_size = 0.1)
print("Finish preparing data.")
```

```python
model = ClassColorNet(n_class = n_classes)
model.fit(
    X_train, ab_train, y_train, X_val, ab_val,
    n_epoch = n_epoch, class_weight = class_weight,
    # load_path = weight_save_path,
    patience = patience, draw = True
)
```

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np




num_test = 3
fig = plt.figure(figsize = (24, 24), dpi=30)
for _ in range(num_test):
    ind = np.random.choice(X_test.shape[0])
    colored = model.colorize(X_test[ind:ind+1])[0]
    img = cv2.cvtColor(colored, cv2.COLOR_BGR2RGB)
    fig.add_subplot(num_test, 3, 3 * _ + 1)
    plt.imshow(X_test[ind].reshape((128, 128)), cmap='gray')
    
    fig.add_subplot(num_test, 3, 3 * _ + 2)
    plt.imshow(img)
    
    fig.add_subplot(num_test, 3, 3 * _ + 3)
    original_lab = np.concatenate((X_test[ind], ab_test[ind]), -1).astype(np.uint8)
    original_bgr = cv2.cvtColor(original_lab, cv2.COLOR_Lab2BGR)
    original_rgb = cv2.cvtColor(original_bgr, cv2.COLOR_BGR2RGB)
    plt.imshow(original_rgb)

    

plt.show()
```
